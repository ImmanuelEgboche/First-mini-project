```python
import pandas as pd 
import numpy as np
```

```python
dataframe = pd.read_csv('british-library-books.csv')
to_drop = [
    'Edition Statement', 
    'Corporate Author',
 'Corporate Contributors',
 'Former owner',
 'Engraver',
 'Contributors',
 'Issuance type',
 'Shelfmarks',
 'Flickr URL'
]
dataframe.drop(columns=to_drop, inplace=True) #WE AND inplace=True TO AFFECT THE DATASET
dataframe.head(15) #HEAD GIVE U THE FIRST 5 VALUES WITH NO ARGUEMENT
```

```python
dataframe['Identifier'].is_unique
```

```python
dataframe.set_index('Identifier', inplace=True) #WE ARE UPDATING THE OLD INDEX TO 'Identifier'
dataframe.head()
```

```python
dataframe.loc[206:, 'Date of Publication'].head(10)
```

```python
col = 'Date of Publication'
for i in range(len(dataframe[col])):
    current = dataframe[col].iloc[i]

    if (type(current).__name__ == 'float'):
        dataframe[col].iloc[i] = np.nan
    
    elif '?' in current or '.' in current:
        dataframe[col].iloc[i] = np.nan
    
    elif '[' in current and not ']' in current:
        dataframe[col].iloc[i] = np.nan
    elif ']' in current and not '[' in current:
        dataframe[col].iloc[i] = np.nan
    elif current.startswith('['):
        dataframe[col].iloc[i] = current[1:5]
    elif '[' in current or '-' in current or ',' in current:
        dataframe[col].iloc[i] = current[0:4]
    print(dataframe[col].iloc[i])
```

```python
dataframe[col] = pd.to_numeric(dataframe[col])
dataframe[col].dtype
```
